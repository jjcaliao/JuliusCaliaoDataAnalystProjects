# **Drop missing values from columns that contain pandas object datatype**

```python
df_nullvals = df[df.isnull().any(axis=1)]

df_dropna_obj = df_nullvals.dropna(subset=['Campaign Name', 'Placement', 'Date'])

df_dropna_obj
# df_dropna_obj.shape
```

# **Replace missing values from columns that contain numbers**

```python
df_fillna_num = df_dropna_obj.fillna(df.mean())
df_fillna_num = df_fillna_num.astype({"Impressions": int, "Clicks": int, "Cost": int, "Video Completions": int, "Find A Dealer Conversion": int, "Online Brochure Download": int})
df_fillna_num
```

# **Now clean whole dataset**

```python
df_cleaned = df.dropna(subset=['Campaign Name', 'Placement', 'Date'])
df_cleaned = df_cleaned.fillna(df.mean())

df_cleaned = df_cleaned.astype({"Impressions": int, "Clicks": int, "Cost": int, "Video Completions": int, "Find A Dealer Conversion": int, "Online Brochure Download": int})

df_cleaned

# verify that there's no null values
# df__cleaned_nullvals = df_cleaned[df_cleaned.isnull().any(axis=1)]
```

# **Count the errors in the dataset**

```python
error1=df_cleaned["Campaign Name"].str.contains("S-Cla_").sum()
error2=df_cleaned["Campaign Name"].str.contains("Merc_").sum()
error3=df_cleaned["Campaign Name"].str.contains("Dis_").sum()
error4=df_cleaned['Campaign Name'].eq('CY21_Q1_LOC_BE_Smart_Fortwo_ProgrammaticVideo_Prosp').sum()

sum_errors=error1+error2+error3+error4
print("There are",sum_errors,"errors in the dataframe contained in dataset_errors.csv")
```

# **Use replace method to correct errors**

```python
for index, row in df2.iterrows():
    df_cleaned['Campaign Name'] = df_cleaned['Campaign Name'].replace({row['Campaign error name']: row['Campaign correct']})

df_cleaned
```
