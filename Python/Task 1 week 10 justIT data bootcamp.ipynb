# **Check if there are any missing values and clean them out**

```python
#prints the whole dataset
df.to_string

#shows null values
#df.isnull()

#use regex to remove commas in Profit and Sales
df['Profit'] = df['Profit'].replace(',', '', regex=True)
df['Sales'] = df['Sales'].replace(',', '', regex=True)

df=df.astype({"Profit": float, "Sales": float})
```
